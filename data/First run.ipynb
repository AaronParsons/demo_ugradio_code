```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib notebook
```

```python
npz = np.load('out001.npz')
print(npz.files)
```

```python
data = npz['data']
sample_rate = 2.2e6 # Hz
time = np.arange(data[0].size) / sample_rate
```

```python
plt.figure(figsize=(4,2))
plt.plot(time * 1e3, data[0])
plt.xlim(0, 0.05)
plt.xlabel('Time [ms]')
plt.ylabel('Volts [arbitrary]')
plt.title('Mixer Setup 1')
plt.grid()
plt.show()
```
